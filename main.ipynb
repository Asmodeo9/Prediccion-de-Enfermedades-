```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Datos de ejemplo: años y tasas de incidencia de varias enfermedades
years = np.array([2021, 2022, 2023, 2024])
disease1 = np.array([16.241, 18.012, 17.739, 4.531])  # Incidencia de Enfermedad 1 (VIH en mexico)
disease2 = np.array([7, 3.177, 7.626, 2.276])   # Incidencia de Enfermedad 2  (influenza en mexico)
```

```python
def fx(x1, coef):
    fx = 0
    n = len(coef) - 1
    for p in coef:
        fx = fx + p * x1**n
        n = n - 1
    return fx
```

```python
# Años en los que queremos hacer predicciones
years_to_predict = [2025, 2026, 2027]

# Lista de datos de enfermedades
diseases = [disease1, disease2]

# Iterar sobre grados de polinomios
for i in range(0, 10):
    plt.figure(figsize=[14, 8])
    
    for j, disease_data in enumerate(diseases):
        coef = np.polyfit(years, disease_data, i)
        x1 = np.linspace(years.min(), max(years_to_predict) + 5, 1000)
        y1 = fx(x1, coef)
        
        # Graficar el ajuste del polinomio
        plt.plot(x1, y1, "--", linewidth=2, label=f'Enfermedad {j+1} - Polinomio grado {i}')
        
        # Datos originales
        plt.scatter(years, disease_data, s=100, edgecolor='black')
        
        # Realizar predicciones y agregar a la gráfica
        for anno in years_to_predict:
            p = np.polyval(coef, anno)
            # Añadir predicciones solo una vez para cada enfermedad
            if anno == years_to_predict[0]:
                plt.scatter(anno, p, s=150, c='red', edgecolor='black', label=f'Predicción Enfermedad {j+1} para {anno}')
            else:
                plt.scatter(anno, p, s=150, c='red', edgecolor='black')

            print(f"Para grado {i} y enfermedad {j+1}, año {anno}, la predicción es {p:.2f}")
    
    plt.title(f"Predicción de Incidencia de Enfermedades para grado {i}")
    plt.xlabel("Años")
    plt.ylabel("Incidencia")
    plt.legend(loc='best')
    plt.grid(True)
    plt.tight_layout()
    
    # Guardar la gráfica (opcional)
    # plt.savefig(f"prediccion_grado_{i}.png", dpi=300)
    
    plt.show()
```

```python
anno = 2025
grado = np.arange(0, 30 + 1, 1)  # De 0 a 30 de 1 en 1

# Listas para almacenar resultados
aproxi = []
mse_values = []

for i in grado:
    coef = np.polyfit(years, disease1, i)
    p = np.polyval(coef, anno)
    aproxi.append(p)
    
    # Para el cálculo de MSE
    y_pred_vec = np.polyval(coef, years)
    mse = np.mean((disease1 - y_pred_vec) ** 2)
    mse_values.append(mse)
    print(f"Para grado {i} El MSE es: {mse:.4f}")

# Convertir listas a arrays para la visualización
aproxi = np.array(aproxi)
mse_values = np.array(mse_values)

# Gráfico de grado del polinomio vs predicción
plt.figure(figsize=[14, 7])
plt.subplot(1, 2, 1)
plt.title("Grado del Polinomio vs Predicción")
plt.plot(grado, aproxi, "--", linewidth=3, color='red')
plt.xlabel("Grado del Polinomio")
plt.ylabel("Predicción para 2025")
plt.grid(True)

# Gráfico de grado del polinomio vs MSE
plt.subplot(1, 2, 2)
plt.title("Grado del Polinomio vs MSE")
plt.plot(grado, mse_values, "--", linewidth=3, color='blue')
plt.xlabel("Grado del Polinomio")
plt.ylabel("MSE")
plt.grid(True)

plt.tight_layout()
plt.show()
```
